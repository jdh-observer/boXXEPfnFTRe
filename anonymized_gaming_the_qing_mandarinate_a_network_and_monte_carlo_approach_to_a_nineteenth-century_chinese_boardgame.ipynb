Qing dynasty (1644-1911), Bureaucracy, Boardgame, Graph Database, Monte Carlo

How did prospective officials during the late Qing dynasty (1644-1911) learn about and perceive their career options? Our study explores the historical boardgame Mandarin Promotions (Shengguantu éå®å). As a career race played with dice, the game resembles Milton Bradley's famous nineteenth-century Checkered Game of Life. However, as it closely mirrors the actual Qing personnel system, Mandarin Promotions is much more complex. We study a late nineteenth-century version reproduced with playable instructions by Puk Wing Kin in 2010, where players hasten from 22 starting positions through 412 positions in 66 departments to reach 18 ranked final positions. We have combined a network approach using a Neo4J graph database and a Monte Carlo simulation approach developed in Python in order to answer three questions: How meritocratic was the imagined official career for the various starting positions which could be broadly categorized into hereditary privilege (including Manchu heritage), the famed imperial examinations, military merit and patronage, and what we call âincipient professionalismâ? Secondly, which were the most pivotal positions for a career to the top echelons of the bureaucracy? Finally, given that office purchase is permitted by the game rules (as in the actual Qing bureaucracy) and that it is the only aspect of the game that requires strategy, we ask if purchasing a higher-ranked position early in the game would be a beneficial strategy. Both methods, but especially the simulation of monetary gains in 100,000 playthroughs, revealed the hidden biases of the Qing's monarchic system which defy common perceptions about a rational meritocratic bureaucracy. Hereditary privilege won big, examination enjoyed modest gains, all other paths lost. Modelling the career system as a network showed the centrality of particular positions, especially that of prefect. Comparison with both statutory regulations and studies of the Chinese Government Employee-Qing database provided two insights: First, the gameboard follows pre-1843 regulations, which confirmed its original creation in 1840, the year on the only surviving board that has a date. Secondly, we interpreted the preeminence of the prefect as an expression of the aspirations of examination graduates. Finally, our simulation of various purchase strategies showed that in most circumstances purchase did not reward the player if monetary gain is considered. In conclusion, our playful engagement with the Qing mandarinate did yield new insights into how contemporaries navigated the world of career-making in a diverse yet monarchic society.

## Intro to the game and research questions

The Qing (1644-1911) mandarinate, though comparatively small by the size of the population it once governed, is among the best documented governing elites in early modern history thanks to its extensive production of official biographies and directories. Quantitative sociological career studies, which began with random sampling of these materials in the 1950s and 1960s, have recently seen a new surge with the publication of large reprint collections and the building of huge databases, especially the Chinese Government Employee Database-Qing by the Lee-Campbell research group in Hong Kong (<cite data-cite="14298532/HNTAEK5M"></cite>). Our paper, while extensively relying on the research findings coming out from the group, will not try to replicate their work. Instead, we are looking at contemporary perceptions of the Qing mandarinate expressed in a popular board game. The game, called Mandarin Promotions (Shengguantu éå®å), has a venerable ancestry going back to the Tang dynasty (618-907) and, in numerous adaptations, enjoyed lasting popularity until the end of the Chinese monarchy (<cite data-cite="14298532/M6PE99IT"></cite>; <cite data-cite="14298532/9RRAW3BJ"></cite>; <cite data-cite="14298532/9QAPAB93"></cite>). The game has been described as a game of power and money comparable to Parkerâs Monopoly (<cite data-cite="14298532/VEI5X3ZT"></cite>). We argue that the career game resembles more closely Milton Bradleyâs famous nineteenth-century Game of Life or the more decontextualized Snakes and Ladders. They share a common origin in the Indian Game of Knowledge (JÃ±Äna ChaupÄr, also spelled Gyan Chaupar) as well as its didactic purpose (<cite data-cite="14298532/TQSWVUUR"></cite>). However, with its 434 positions and multiple starting and final points Mandarin Promotions is much more complex than these well-known board games. It is thick in social and historical context, since the game designer (in a colophon printed on the board) claimed that the game closely mirrors the reality of the Qing bureaucracy. The game has been rediscovered recently, and Puk Wing Kin has produced a playable introduction (<cite data-cite="14298532/5Z849G6F"></cite>) of a late nineteenth-century version which forms the basis of our exploration. We will compare a network approach using a Neo4J graph database and a simulation approach developed in Python. We hope to find out what kind of expectations for a bureaucratic career the game raised in the players and how they corresponded to the bureaucratic realities of the day. This introduction will describe the rules of the game and our research questions. Section 2 introduces our methodologies. In the following four sections, we compare the two methods in answering our exemplary research questions. Our findings confirm that the game is closely modelled after the statutory appointment procedures, albeit of the early rather than the later nineteenth century. Mandarin Promotions reflects the aspirations of the literati for an examination career, while clearly showing a monarchic outlook and acknowledging hereditary privilege.

### The Game

Mandarin Promotions has been circulating, at least since the early seventeenth century, in two versions. One plots the entire system of civil and military offices and is played with six dice. This board has first been introduced in Europe, albeit incompletely, by Thomas Hyde in 1694 (<cite data-cite="14298532/EK62UNFP"></cite>, 219-229). We are dealing here with the smaller version limited to civil officials played with four dice (<cite data-cite="14298532/5Z849G6F"></cite>). Mandarin Promotions game boards were ephemera printed on paper. Therefore, not many boards have been preserved. From the surviving boards it becomes clear that the game board was continuously adapted, with versions representing historical as well as contemporary bureaucratic systems. For example, if we compare Thomas Hydeâs six-dice board with a surviving Ming dynasty version preserved in the WolfenbÃ¼ttel library (<cite data-cite="14298532/4UJ6E3QS"></cite>, 91), the former added official titles of the new Manchu dynasty. We have seen three nineteenth-century boards for civil officials (four dice), one 1840 board (the only of the three that is dated) held at the Bodleian library in Oxford () and two boards from the later nineteenth-century, one reprinted from the Cleveland University Library (<cite data-cite="14298532/9QAPAB93"></cite>), the other one, which carries the imprint of a Shanghai publisher (hence Shanghai board), has been reproduced and studied by Puk Wing Kin (<cite data-cite="14298532/5Z849G6F"></cite>). All three boards share most features as well as most of the rules and the anonymous preface printed in the center of the board (the date is removed on the later boards). They are also similar to a board described by Culin, played by Chinese immigrants in the USA in the 1890s (<cite data-cite="14298532/CZ5ZCA69"></cite>, 504-507). This shows that by the nineteenth century the game was relatively standardized. However, the Cleveland and Shanghai boards added a box for office purchasers (to be explained below). A new version was created as late as in 1910, when the monarchy began to reform its administrative system, but it soon became obsolete with the revolution one year later (<cite data-cite="14298532/KJK3I95E"></cite>).

### The Game Board

Our exploration will be based on the Shanghai board reproduced by Puk Wing Kin who has also provided playable instructions. Before we introduce the rules and the research questions, three characteristics of the game have to be mentioned. First, the Mandarin Promotions board plots 434 bureaucratic positions in 66 departments which demarcate the space for career making. However, different from the common conception of the mandarinate as the bedrock of meritocracy, the social vision of the game is one of monarchy in which the emperor assumes the role as the center of arbitrary power. In the games derived from the Indian Game of Knowledge (Checkered Game of Life or Snakes and Ladders) the goal is at the end. In Mandarin Promotions all paths lead to the center. This appears similar to the European Game of the Goose, but in Mandarin Promotions there are many more paths and none of them is linear, increasing the arbitrariness and, hence, suspense of the game. Mandarin Promotions also is neither a game of salvation nor a mere race game. It is an imperial metaphor. In the earlier six-dice version, the center depicts the gate of the forbidden city, but in the nineteenth-century boards, the square in the center contains the rules of the game. It is also the place into which a bowl is placed to throw the dice. The throne is conceived as the center of arbitrary power, while the bureaucracy is arranged in concentric circles around it, starting from courtly titles and offices, to the metropolitan bureaucracy, to the provincial administration, the civil service examination, and the system of promotions and penalties on the outermost circle.

Secondly, similar to life games, Mandarin Promotions is played by dice with little room for strategy or choice. The next step following certain dice rolls is always spelled out in words below each position. Dice combinations are not counted numerically but numinously. They are imbued with value: VIRTUE (Double-4), TALENT (Double-6), MERIT (Double-5), MEDIOCRE (Double-3), MALLEABLE (Double-2), CORRUPT (Double-1 can be cancelled by a 4). One could argue that the dice combinations signify merit in a countable way, contradicting the above contestation that the vision of the game is less rational-meritocratic than expected. However, merit is defined by fortune (dice) alone (This is also the case in the Tibetan game of Liberation. See above.). There are also âluckyâ dice without numinous names, particularly four equals (*quanse* å¨è²) and 3-4-5-6 (*chuanhua* ç©¿è±, garland), and they seem to ensure the best career moves, although this proposition has to be tested in our simulation.

Thirdly, in Indian and European life games, there is only one starting position, and it is usually the first square or field within the game, often designated âbirthâ or âgoâ (as in Monopoly). There is also only one final position. In Mandarin Promotions, players start from outside the game (the term â*juwai* å±å¤â, lit. âoutside the gameâ, today also denotates the âoutsider,â *juwairen* å±å¤äºº, in general). The first roll of the dice determines from which of the fifteen starting positions the player will start. In addition, there are a number of hereditary titles of nobility attained with âluckyâ (all equal) dice only. The starting positions correspond to the real-life notion of â*chushen* åºèº«â, lit. provenance. These are qualifications which lock  the holder into certain career paths. We can distinguish roughly five types of provenances: examinations, heredity and imperial grace, practical merit and patronage, incipient professionalism, and finally purchase. Out of 15 chushen positions, only three are related to the famous civil service examinations. This is different from the seventeenth-century boards where the examinations were the dominant path into office.

### The Rules

1. The game is played by an unspecified number of players. Our simulation will assume four players. For the first roll of dice, special combination rules apply which determine the starting positions, of which there are 21. Six of these are titles of hereditary nobility and are reached with âlucky diceâ (four equals from 4 to 1). 4, not 6, is the luckiest number as is common in most Chinese dice games. The remaining 15 are the chushen positions described above. 
2. Upon each further roll of dice, players move their primary game piece from position to position according to the instructions printed on the board. From each position five to six other positions can be reached. Regular positions correspond to steps in the civil service examinations or to ranked positions in the bureaucracy. However, out of the 434 total positions, 123 are reserved for a secondary game piece, that is placed on them while the primary piece remains in its position. In the next round, the instruction that is prescribed by the secondary position is applied to the primary game piece. The secondary piece is then withdrawn. These temporary positions represent three institutions in the Qing bureaucracy:
 1. Performance reviews taking place every three years and penalties based on impeachments (49),
 2. Commissions, i.e. temporary work assignments while the main position or rank is on hold (38),
 3. Special honorary rewards and titles meted out by the emperor (36).
3. The goal of the game is to rise in rank-class and move to retirement at the highest possible rank-class. There are 19 rank-classes, with 1a being the highest and unclassed the lowest. Players are sent into retirement from their main position (rewarded retirement, *he* è³ or *dahe* å¤§è³) or from the performance or penalty position (honorable retirement, *yugao* äºå, or dishonorable retirement, *xiuzhi* ä¼è´). The player moves his game piece into the section Rank Review (*pinjikao* åç´è). However, questions remain. What exactly counts as âwinâ? Does being first to retire also mean earning the most money? If these two are not the same, would players prefer earning money or finishing with high status? These questions have to be tested in our simulation.
4. According to the rules, the game ends when every player has reached the retirement section. However, this rule is not practicable, because the game could go on for a long time with only one player left who continues to receive payouts from the pool. Once the pool is empty, co-players who already âretiredâ should no longer be required to replenish it, in analogy to rule 5.C.a. below. Hence, in our simulation, we have decided that a game is considered to be finished when either all players have moved into retirement, or, with only one player left, the money in the pool has run out.
5. Every player puts 100 chips into the pool at the beginning of the game. Money is paid at various stations in the game: 
    1. From the pool to the player
        1. Rewards: At several positions the player receives quite large rewards from the pool. These are mostly titles of nobility and special imperial rewards.
        2. Allowances for commissions
    2. From the player to the pool
        1. Fines
        2. Office purchase
    3. Gifts from other players
        1. Congratulation gifts, for example the six titles of nobility on the first round, succeeding among the top-three holders of the highest examination degree, and upon rewarded retirement. Players who have already retired do not have to pay these gifts.
        2. Meeting-gifts, when two players land in the same government department. This is paid by the lower-ranking official to the higher-ranking one.
    At the end of the game, any remaining pool money is first shared equally between the players. Then lower-ranking players pay out higher-ranking ones by counting the difference in rank class. For example, if one player ends his career ranked 1b and the other one 5b, the rank difference is 8. At a rate of 5 chips per rank-class, the lower-ranking player pays 40 chips to the higher-ranking one. 
6. Office purchase is  the only mechanism that breaks the above-described flow of the game. The written rules printed on the 1840 Oxford board already allow office purchase up to a certain rank, which corresponds to the laws governing legal office selling at the time, but no specifics concerning the procedure and advisability were given. Only two functions of office purchase are mentioned, namely to bail out from an undesirable starting position (*chushen*) or from a penalty dismissal. The most undesirable starting positions of all were students of astronomy and medicine, a clear expression of a value system that cherishes noble status over professional work. Interestingly, the only two major changes in the written rules from the 1840 board to the two later boards meant a hardening of the stance towards the doctors and astronomers (One rule stipulated that office purchase for bailout was only allowed in the second round, that is immediately after the player had landed on the respective chushen position, but not at a later point. The other rule change excluded them from titles of nobility high up in their career if they had missed the bailout and proceeded with the game.). Late nineteenth-century game boards also created a new category of expectant official, a sort of holding for office purchasers before they were appointed, then again by dice, to full positions. In both ways, office purchase is the only agency in a game that was otherwise entirely determined by chance. Modern recreations of the game in Hong Kong discourage or ban office selling, because it destroys the suspense of the game (Puk Wing kin, the historian, encourages his readers to ban office purchase, even though he faithfully explains the original rules that allow it (<cite data-cite="14298532/5Z849G6F"></cite>, 29). Pan Guosen who has made major changes to the game follows the Oxford chart and eliminates the æç­åè£ block (<cite data-cite="14298532/4E7IWGZF"></cite>, 87-88). He still allows purchasing a reinstatement into previous office (*juanfu* æå¾©) after a penalty dismissal (p. 97). However, from the historianâs perspective the insertion of the new category of expectant officials by purchase represents the expansion of office selling in the early nineteenth-century (<cite data-cite="14298532/BIJ42UER"></cite>). Our simulation will therefore try to determine under which circumstances office purchase would be beneficial, considering both the purchase price, set in the game as 5 chips per rank-class, and the final winning opportunity.

### Research Questions

Our exploratory digital study hence pursues three questions.

First, how meritocratic was the career path? Given that there are fifteen *chushen* positions and six titles of hereditary nobility, did every of these starting positions have the same chance to win the game, both in terms of speed to get to the highest position and in terms of money? How do the celebrated civil service examinations fare in this process?

Secondly, how central are certain official positions for the career? Are some positions more pivotal than others? Related to this question is the problem of historical accuracy and representation. The anonymous preface claims authenticity: âBased on the Collected Statutes (Huidian), we have made a comprehensive table of official ranks, complete for Manchu and Han officials, and differentiated between the regular and irregular paths. It is not just a game, but faithfully lets the distinctions of official ranks and the understanding of their qualifications be clearly spread out in one table, in order to be of great service.â(<cite data-cite="14298532/5Z849G6F"></cite>, 22; for a full, but slightly different translation see: <cite data-cite="14298532/FW2IZMHP"></cite>, 76) Now, official promotions in the bureaucracy were highly regulated by the Statutes of the Board of Personnel. Comparison will be made with the quantitative research coming out of the Lee-Campbell group as well as with the statutes. Since statutory career steps in the regulations changed with time, such a comparison may offer clues about the gameâs historicity and the time reflected in it. 

Third, we want to know which career path is the best? Is the fastest path always the best monetarily? How can the player influence the outcome of the game? As stated above, strategy is not part of the game design. The only agency of the player is to purchase a position. This is what the preface refers to as the âirregular path.â (The regular path (*zhengtu*) into office holding is hereditary privilege or success in the imperial examinations, the irregular paths (*yitu*) are office purchase, military merit and patronage recommendations.) It was irregular but neither illegal nor illegitimate. Can purchase of office propel oneâs career AND yield monetary reward at the same time? In other words, does purchase pay?

### Data Modeling in Neo4J

A network approach may facilitate shortest path searches and centrality measures. Modelling the game as a graph is justified by the structure of Mandarin Promotions where different career paths cross the same positions. This creates a network in which positions can exhibit different degrees of centrality. By modelling the data to create a network rather than a tree where each branch would correspond with a career path, possibly visiting the same position multiple times, we hoped that it would be possible to understand how different positions in the bureaucracy were linked to each other and which were the most important positions. We decided to use a Neo4J database which offered several advantages. First, different from simply feeding CSV files into a network tool like Gephi, Neo4J allows to store an unlimited amount of properties on both nodes and relationships. This allows to maintain the full historical information. For example, we can store the dice value both in its traditional form and as a numeric value for computation. The properties also provide a place to list the money earned or paid during the game. Neo4J also is fully compatible with UTF-8 Chinese characters. Secondly, new relationships can be created by executing custom scripts, which is especially important for the rule-bending properties of the temporary positions (commissions, performance reviews, rewards and punishments) which cannot be simply entered from a tabular database (CSV). Third, Neo4J promised to handle directed graphs. Most network tools are primed for undirected graphs. Although Neo4J is a graph database rather than a network analysis tool, it also provides such tools and constantly improves them. Finally, this graph database from Sweden has a large presence on the internet. It is free to use for non-business projects, and provides teaching tools for learning. Disadvantages of using the network approach are, first, that it requires to define temporary positions as nodes, although they are rather (sub-)states of the main node. Secondly, it is difficult to simulate multiple players which would be necessary to calculate monetary transactions.  We solved this by using different tools to investigate different research questions. Third, Neo4J has not been designed for the purpose of game analysis and has to be bent to fit. Since Neo4J is an evolving software, versioning and frequent changes make handling somewhat complicated. 

The first and most important step in applying the network approach was to devise a data model. As stated above, the Shanghai board of Mandarin Promotions reprinted by Puk Wing Kin has 434 positions in 66 compartments. These form the nodes of the graph. From the CSV files, we have entered all nodes with their Chinese names only, but English translations from Brunnert & Hagelstrom could be easily added (<cite data-cite="14298532/U6KCKNKQ"></cite>).

#### The Nodes

Positions on the game chart are distinguished by the attached dice rules into two broad categories. The first category consists of nodes that are followed by basic rules of movement to the next node. We designate those as *uniquenodes*. Of 434 positions, 338 positions are uniquenodes. In the graph, each can appear only once. From the perspective of the graph or game, they behave in similar ways, but for their bureaucratic function, we have still distinguished them by adding the labels Start, Final, Exam, and Ranked. Start, Final, and Exam are self-explanatory. Ranked positions are official tenures associated with one of the 19 rank-classes. All uniquenodes have as properties the compartment on the game chart, mostly the government department the position belongs to, their location (metropolitan or provincial), and their rank-class where appropriate. We have invented one uniquenode that is not on the game board, namely that of that of the game outsider (*juwai baiding* å±å¤ç½ä¸) who enters the game with the first roll of dice on one of the 21 starting positions.

The second category of nodes consists of temporary positions. Instructions under these nodes read something like âIf you roll âvirtueâ then exercise one âvirtueâ and one âmeritâ moveâ and refer to the movements of the primary game piece. These nodes, which we refer to as concurrentnodes, are entered into the system with the labels Performance, Commission, and Honorific. Performance nodes represent performance reviews, most prominently the triennial reviews for provincial officials (C12, *daji* å¤§è¨) and for officials in the capital (C10, *jingcha* äº¬å¯). Commissions are temporary assignments for metropolitan officials, as explained above. Honorifics are special awards (C64, *teâen* ç¹æ©) like the peacock-feather plume decorations (P413 *hualing* è±ç¿) or titles of nobility (C63) like marquis (*houjue* ä¾¯çµ). ### Monte Carlo Method

While the approach of modeling the game as a network can be used to gain insight into several aspects of its design, there are also some limitations that are hard to overcome. Specifically, there are mechanisms in the game that need to take into account the precise path that a player has taken before ending up in a certain position. First, to make accurate statements about monetary aspects of the game, each transaction with the pool or in between players has to be recorded. Secondly, there are a few special rules in the game that limit access to certain positions depending on the background of the player, that is, whether or not they ever held specific positions earlier in the game. While in theory, it would be possible to account for these in a network, by defining each state as a tuple similar to the approach for concurrentnodes described above, in practice this would lead to an unacceptable blow-up in the state space.

Another possibility for the analysis of a game would be deriving a system of mathematical equations that give an explicit solution to the various probabilities that one might be interested in. This has been done e.g. for the Game of the Goose (<cite data-cite="14298532/I6PTQ8ZZ"></cite>), however, this is already quite involved for the much simpler Game of the Goose without even taking monetary aspects into account.

Hence, we have decided to pursue a straightforward yet effective approach, using a Monte Carlo Method. Monte Carlo Methods have been used extensively to study systems that are too complex to derive an explicit solution. They work by simulating the system for a large number of times, such that by the Law of Large Numbers, the observed distribution of outcomes will converge to their expected values.

In practical terms, in order to conduct the Monte Carlo experiment, we implemented the rules of the game in Python, using the existing data in the CSV datafiles as a basis. While the regular flow of the game was relatively easy to implement given these, the simulation becomes much more difficult when also considering special rules that only apply in certain circumstances, such as a few positions being restricted to persons coming from a particular background, or the game mechanisms for astronomers and physicians. Furthermore, while the course of the game is mostly decided by the roll of the dice, there are a few elements of choice, especially the option to purchase an office, for which we had to decide on and implement realistic strategies of how players would behave when they have the opportunity to influence the run of the game. In order to fit these different strategies into a Monte Carlo based system of analysis, we simply run the experiment with different strategies and compare the results. 

## Do all provenances have equal opportunities? Network vs. Simulation 

Our first question is whether all fifteen chushen provenances have an equal chance to reach the highest rank class. This relates to the question of equal opportunity, if only in a game, not in life. If opportunities are highly unequal from the beginning, the game loses suspense. As we have already seen in the case of Astronomers and Physicians, the anonymous game designers realized this problem and strongly encouraged the two starting positions of âStudent of Astronomyâ and âStudent of Medicineâ to bail out. In this section, we try to compare career prospects using first a shortest path search in Neo4J and second a simulation over 100,000 games with four players each. The shortest path analysis performed on the network model of the game can be used to compute those career paths that involve the least number of intermediate positions between two nodes of the network. In order to better understand the opportunities of the different chushen in relation to each other, we can e.g. compute the shortest paths from the various starting positions to the highest final position Guanglu Daifu (P417 åç¥¿å¤§å¤«, 1a). Intuitively, we expect that this corresponds with a high chance of a player starting from that position to retire quickly in a high position. We also anticipate that a fast way to retirement at the highest position could mean that the player has a good chance at making a profit from the game. This is because players going into retirement can only receive âcongratulationâ money from those that have not finished yet. Moreover, the first player to retire can expect a higher payout at the end, if co-players are retiring at a lower rank.

However, in the real game, players will mostly not proceed along the shortest path. Nor will they all retire at rank 1a. Hence, the shortest path analysis could also be misleading in terms of the expected outcomes of the game. For example, a starting position (*chushen*) may have a shorter shortest path to a good retirement position than another, but on average, the paths originating from it are longer. After all, for any given path, the probability of a player to stay on it decreases exponentially with the length of that path. Furthermore, when it comes to the money the players can gain, a longer path might actually be beneficial, because having visited more positions, especially at the top levels, might mean that the player has more opportunities to get rewards from the pool or gifts from other players. Also, even when ignoring the special lucky dice combinations, not all paths are equally likely, since a corrupt dice will be canceled by a four, so it has a lower probability of being the next move from any position. Finally, as mentioned above, one aspect of Mandarin Promotion is that it has multiple distinct final positions, so ideally, a shortest path analysis should take all of those into account, which, however, is difficult to do.

Using the simulation approach, we can realistically compute how the expected outcomes of the game would present themselves to players. A potential downside of this type of analysis is that, since it can take a large number of games played until the observed data converges to the expected value (<cite data-cite="14298532/I6PTQ8ZZ"></cite>, 144), and game design was not necessarily informed by deliberate stochastic considerations (<cite data-cite="14298532/TTRIXB4J"></cite>), the âexpected valueâ in the modern sense of the word might not correspond to the actual considerations of the designers and the subjective experiences of players of the game in historic times. 

### What are *chushen* provenances?

As mentioned above, Mandarin Promotions technically starts from a position outside the game. Commoners (*baiding* ç½ä¸) have no stake in the game and remain âoutsidersâ (*juwai*, literally âoutside the gameâ, a term still used to describe outsiders in general), unless fortune (the initial roll of dice) gives them provenance (*chushen* åºèº«). Historically, having a chushen (provenance) was the central precondition for any entry into the ranks of officialdom. Oddly, the term has been revived by Mao Zedong in the 1950s to denominate hereditary class background (<cite data-cite="14298532/UD5CABPH"></cite>).

Surviving game boards from the Ming dynasty and the early Qing (six-dice full civil and military official charts) still proceed from the assumption that only the imperial examinations (both the civil and military variety) provide a viable chushen provenance that leads to office-holding. As we have already seen, this assumption no longer holds true in the nineteenth-century game boards when the Manchu rulers had long put their stamp on elite formation and the Jiaqing Emperor (1796-1820) had expanded office selling.

The chushen positions listed in the first two compartments (C1 åºèº«, C2ä¸çµ) can be roughly divided into the following categories:

1. As can be seen from [Table 1](table-chushen1-*), hereditary privilege dominates. These include bannermen, i.e. members of the ruling Eight Banners (P7, P12, mostly Manchus), descendants of Confucius (P16, with the best possible dice combination, four 4s), and lower nobility (P1, P17-P21). Such titles are usually given to the sons of high-ranking and meritorious officials. The titles of nobility P17-P21 are translation of Manchu terms bestowed upon military heroes and their posterity. Even though this chart pertains only to civil careers, heredity and military merit appear in conjunction and were highly valued in the Manchu-ruled universe. 
2. Several more chushen positions honor merit outside the imperial examinations, again with an emphasis on military merit (P2, P4, P5). Patronage played an important role in gaining the recommendations required for advancement in this category.
3. The famous civil service examinations are represented in P3, P6, P10, P11, P15. However, in reality, P3 was often and P11 was mostly venal, while P6 (è©ç§, special palace examination) was a rare event that required a recommendation to participate (patronage), leaving the student (P15) and licentiate (P10) as pure examination positions. There may be a good deal of irony in the game designerâs choice for the next round: For a Manchu Banner School Student (P12), being thrown back to becoming a Licentiate, the most sought-after examination degree position (Double-4, âvirtueâ), is the worst outcome. It also speaks to the carnivalesque quality of the game that the roll of dice also determined the ethnicity of the player. 
4. The fourth category is incipient professionalism. Apart from the maligned Astronomers and Physicians, this category includes clerks. Together, they start from the worst dice combinations (Triple-2, Triple-1, Double-3, Double-2). In another spark of irony, only the student (P15), i.e. the unsuccessful candidate for licentiate status (P10) fairs worse (Double-1). Lowly bureaucrats below the level of ranked officials, the clerks could legally rise into low-ranking officialdom. The question is whether they had good chances to rise to the top.

### Network Results

We are using shortest path searches in Neo4J to find out whether all fifteen *chushen* positions have an equal chance to reach the highest-ranking Final (*guanglu daifu* åç¥¿å¤§å¤«, 1a). Neo4J provides sophisticated pathfinding algorithms, but we have so far found them rather unsuited to our needs. For one, they provide only one path, but we would like to know how many different paths exist and what they are. Secondly, they cannot work with negative weights, while monetary fines in my database are expressed as negative numbers. There is a workaround, but it was more complicated than the following solution.

The following table ([Table 2](table-allShortestPaths-*)) is computed from simple allShortestPaths (source-target) searches using CYPHER. The paths are unweighted, but gains are computed from rewards received from the pool, fines paid to the pool, and gifts received from the other players. In this way, the paths can be sorted from lowest to highest gain. Money from players has been factored in the results based on the assumption that there are four players, so that the gift amounts in the rules have been multiplied by three.

    MATCH(a:Start {name: 'è­ç'})
    MATCH(b:Final {name: 'åç¥¿å¤§å¤«'})
    MATCH p=allShortestPaths((a)-[:MERIT*..25]->(b))
    WITH DISTINCT p
    WITH p, relationships(p) AS r
    UNWIND r AS rels
    WITH p, count(rels) AS hops, sum(rels.player_money)*3 AS player, sum(rels.pool_money) AS pool 
    RETURN hops, player, pool, player+pool AS gain ORDER BY gain DESC

### Simulation Results

In order to analyze the different *chushen*, we simply group the data collected in the simulation by the first position on the gameboard that the player arrived at and take in-group averages for the values we are interested in. Since the simulation starts â just as the game â outside the gameboard, and follows the prescribed rules in assigning chushen positions according to dice rolls, in the collected data, the *chushen* positions that are associated with dice combinations that are more likely to occur are overrepresented, while other chushen are underrepresented. Hence, to have a meaningful analysis, it is important to conduct a sufficient number of experiments so that even for unlikely starting positions there is enough data.

Looking at the data in [Table 3](table-chushen-simulation-*) below, that contains the average number of hops until retirement and average effective money for the different chushen, we can first observe that the average number of moves that a player needs to take before reaching a retirement position is much higher than the respective shortest path. The average over all playthroughs in 100,000 simulations with four players is as high as 37, being most heavily influenced by those chushen that one can obtain with a double, i.e. P10-P15. Observe that in the simulation, a move corresponds to one hop in the network, which is not the same as a roll of dice, as some combinations will not lead to any change in the gameboard, while for others, the player can take multiple steps at once.

```python
import re
import functools
import gzip
import matplotlib.pyplot as plt
from matplotlib import font_manager
import pandas as pd

from script.csv_parser import read_csv
from script.statistics import PlayerStatistics

# Since the game code is too long to present here,
# we load existing data from our simulation.
# The game is implemented in script/game.py.
# The data needed to run the simulation is 
# contained in the directory script/data.

stats = []
# The statistical results are contained in files that
# are suffixed with the purchase strategy used (see below).
# For the moment, we will always use the data with 
# no office purchase.
with gzip.open("script/statistics_never.txt.gz", mode="rt") as infile:
	lines = infile.readlines()
    # Each line contains a summary for one playthrough
	print(f"Read input file 'script/statistics_never.txt' containing data for {len(lines)} plays")
	
	for l in lines:
		stats.append(PlayerStatistics.from_line(l))

# Group stats by first position
stats_by_initial_position = dict()
for s in stats:
	stats_by_initial_position.setdefault(s.first_position, []).append(s)

# Function to compare the names of two positions, used for sorting
def positions_cmp(x, y):
	position_re = "C([0-9]+)-.+P([0-9]+)"
	
	match_x = re.match(position_re, x)
	match_y = re.match(position_re, y)
	
	c_num_x = int(match_x.group(1))
	c_num_y = int(match_y.group(1))
	
	if c_num_x < c_num_y:
		return -1
	elif c_num_x > c_num_y:
		return 1
	else:
		p_num_x = int(match_x.group(2))
		p_num_y = int(match_y.group(2))
		
		if p_num_x < p_num_y:
			return -1
		elif p_num_x > p_num_y:
			return 1
		else:
			return 0

stats_by_initial_position = sorted(stats_by_initial_position.items(), key = functools.cmp_to_key(lambda x, y: positions_cmp(x[0], y[0])))

average_pool_money = list(sum(s.money_from_pool - s.money_to_pool for s in ss) / len(ss) for _, ss in stats_by_initial_position)
average_player_money = list(sum(s.money_from_players - s.money_to_players for s in ss) / len(ss) for _, ss in stats_by_initial_position)
average_effective_money =  list(sum(s.money_from_pool - s.money_to_pool + s.money_from_players - s.money_to_players for s in ss) / len(ss) for _, ss in stats_by_initial_position)
average_num_moves = list(sum(s.num_moves for s in ss) / len(ss) for _, ss in stats_by_initial_position)
average_final_rankint = list(sum(s.final_rankint for s in ss) / len(ss) for _, ss in stats_by_initial_position)

# Show overview in DataFrame
pd.DataFrame({"Count": [len(ss) for _, ss in stats_by_initial_position],
              "Average number of hops until retirement": average_num_moves,
              "Average effective money": average_effective_money,
              "Average final rank (1a = 19, unclassed = 1)": average_final_rankint},
            index=[p for p,_ in stats_by_initial_position])
```

The discrepancy between shortest path and average length does not mean however that the length of the shortest path is entirely meaningless for our understanding of the game. Actually, we can see that there is â at least for some of the positions â a correlation between the length of the shortest path and the average number of steps until retirement. Comparing [Table 2](table-allShortestPaths-*) and [Table 3](table-chushen-simulation-*), this can be seen especially clearly for the hereditary positions (P17-P21) that are obtained through lucky dice. At the same time, the simulation data also yielded some results that are unexpected when only considering the shortest paths. In the category of incipient professionalism, the astronomer (P8) who has the longest shortest path to the highest-ranking final state has an average number of steps that is  on par with the average of all playthroughs. This could be due to the special retirement rules for astronomers that will often lead to them not retiring in the highest rank, which is not captured when only searching for shortest paths to the highest final positions. This, of course, is even more true for the students of medicine (P9), who are not even able to reach the highest rank, but are on average very quick to retire.

Sub-official clerks (P13, P14), on the other hand, had longer average lengths over all playthroughs. In other words, on average, they required more hops to retirement. They shared this feature with the Manchu students (P12) and candidates for the lowest examination degree (P15). Now, these four starting positions, equally had shortest paths of 12 hops, but very different numbers of possible shortest paths. The results of the shortest path search and the statistical analysis do not correlate. Therefore, these two tests are still inconclusive as to whether one *chushen* position is more advantageous than the other, especially since being fast to retire is not the only goal of the game To answer this question, we need to turn to calculating their monetary rewards [Figure 8](figure-money-by-starting-pos-*).

## Which positions are pivotal for official careers?

A network graph allows to use centrality measures in order to find out which positions in the bureaucracy were more pivotal than others. Graph Neo4j has two plugins which can answer graph global questions. Graph Data Science is the stronger of the two, and the Graph Data Science Playground allows to apply algorithms without coding them yourself. However, we found it more useful to implement the algorithms ourselves, because the presentation of results can be better adopted to our needs. we have limited the application of algorithms to the relationship type MERIT.

There are three types of centrality of nodes of which two appear useful for our analysis: Degree Centrality, Betweenness, and PageRank. We discuss our results obtained with these three methods below. To check whether these measures correspond with the reality of the game, we also count how often positions are visited in the Monte-Carlo approach.

### Centrality Measures

### Counting Visits in Monte Carlo

To get an idea of how important different positions are using the Monte Carlo approach, we use the most straightforward and simple approach of just counting how many times the positions are visited. The results from simulating the game 100,000 times are presented below ([Table 8](table-visit-count-*)), excluding final positions to get a comparable result with the centrality measures above.

```python
positions_visit_counter = 338 * [0]

for s in stats:
	for i in range(338):
		positions_visit_counter[i] += s.positions_visit_counter[i]

positions_visit_counter = sorted(enumerate(positions_visit_counter), key=lambda x: x[1], reverse = True)
# We use a custom CSV parser that normalizes the spreadsheet data
positions = read_csv("script/data/positions.csv", "PosID,ranktrad,rankint,Name,Type,Department,Location,Remarks".split(","))
# Filter final positions
positions_visit_counter = [{"position": positions[i]["PosID"],
                            "rank": positions[i]["ranktrad"],
                            "location": positions[i]["Location"],
                            "absolute visits": c} for i, c in positions_visit_counter if not positions[i]["Type"] == "Final"]
pd.DataFrame(positions_visit_counter)
```

## Historicity: The Case of Prefects

Given its centrality in the game, it is surprising to find that prefects have not received a lot of attention in historical scholarship. Four bodies of sources have been used in quantitative studies of official appointments: lists of officials in local county, prefectural, and provincial gazetteers (*difangzhi* å°æ¹å¿) (<cite data-cite="14298532/8GWCF962"></cite>; <cite data-cite="14298532/W59CEQHJ"></cite>), official directories compiled by the provinces (*tongguanlu* åå®é) (<cite data-cite="14298532/S8FH7KEH"></cite>), empire-wide official directories compiled in Beijing (*jinshen quanshu* ç¸ç´³å¨æ¸, *juezhi quanlan* çµç§©å¨è¦½) (<cite data-cite="14298532/ZUD52EZ8"></cite>), and (more recently) archival collections of CV forms prepared for imperial audiences (*lÃ¼li* å±¥æ­·) (<cite data-cite="14298532/VGZS5N33"></cite>; This work has been facilitated by the publication of these documents: <cite data-cite="14298532/6MXCA39B"></cite>). Of these sources, only the provincial directories and the CV forms are rich in context, but they tend to be somewhat random and scattered. The most ambitious attempt at gaining complete coverage has been made by the Lee-Campbell group at HKUST with the digitization of the metropolitan directories (<cite data-cite="14298532/HNTAEK5M"></cite>). All of these sources include information about prefects. However, prefects have not been in the center of attention (Only Lee Kuo-chi et al and Wang Zhimin have devoted some space to the prefects. Recently, Peng Peng has used gazetteers to study prefects, but she does not study the appointment process but rather the outcome, using regression analysis, for example, to compare the outcomes of crises for prefects of different background (examination graduates and purchasers). She found no difference.). This has changed with a recent study by Hu Heng et al. which is part of the excellent and growing body of research that is being produced by the Lee-Campbell group and on the basis of their work (<cite data-cite="14298532/NMH52TDZ"></cite>).

In the Qing administrative system, the prefecture is a unit of territorial administration that is between the county and the province. A prefecture has several counties, and a province has several prefectures. Since the city in China did not have an independent administrative existence, prefectural cities often are shared between two counties hand hence the seat of at least three administrators: two magistrates and one prefect. Important prefectural cities could also be the seat of a circuit intendant (a functional and territorial unit above the prefecture) and even the provincial governor if it tripled as seat of two counties, the prefecture and the provincial capital. Considering that official documents, as in most bureaucracies, had to pass up the administrative ladder (the âDienstwegâ), prefects must have had a pivotal role in the flow of information in the bureaucracy. The question is how was this role expressed in bureaucratic appointments?

Hu Heng et al. focus on appointment procedures in the appointments 3403 prefects in CGED-Q (1833-1911). Prefectures had designations of importance. These are expressed in the board game as âvibrant (i.e. lucrative) prefectureâ (*fanfu* ç©åº, P117) vs. âsimple prefectureâ (*jianfu* ç°¡åº, P 118). They also had designations of appointment prerogative, namely: 1. appointment by imperial edict (*qingzhi que* è«æ¨ç¼º, 48.3% of vacancies); 2. appointment upon request of the governor (*tidiao que* é¡èª¿ç¼º, 26.1.% of vacancies); 3. appointment by Board of Personnel (*buxuan que* é¨é¸ç¼º, 25.6% of vacancies). This makes prefectures a perfect case to discuss the balance of power between the court, the Board of Personnel, and the provincial governors.

Hu argues (p. 368) that 43.3% of all vacancies (102) were to be appointed by imperial edict *qingzhi que* è«æ¨ç¼º, and these were located in the most important, connected and wealthy places. This shows that the throne maintained centralized power over the appointment process, while the Board of Personnel was comparatively weak, finding its appointments infringed by both the throne and the governors. 

Hu Heng et al. and Wu Yue have been sparring over the validity of these findings, with Wu Yue arguing mostly against the reliability of the official directories as a source. More often than not, they reflect an abstract appointment to a rank with no guarantee that the candidate actually showed up on the ground (<cite data-cite="14298532/5NCBAEXH"></cite>). Wu Yue also somewhat dampens the idea of a power struggle in the appointment designations. The appointment procedures were highly bureaucratized. The only ways for the governors to break away from these procedures were those not expressed in the official directories, namely sending appointed prefects on commissions and temporary substitutions elsewhere, while filling the vacant posts with their own substitutes. However, from our perspective of an aspirational board game, the murky waters of provincial dealings are not so important. Abstract appointment is all that counts.

## Money and Office Purchase (Simulation)

### The Problem of Office Purchase and Some Made-up Rules

The legal sale of official rank and appointments has a venerable history, going back to the beginnings of the Chinese empire and certainly to the beginning of the Qing dynasty (<cite data-cite="14298532/9UWMJPS2"></cite>). The Qianlong emperor institutionalized the sale of brevet rank (without eligibility of appointment), but used the sale of full rank sparingly. This changed after 1798. In the early nineteenth-century, office selling became a regular experience. The design of the 1840 version of Mandarin Promotions began to reflect this change, as we have seen above. However, before the Opium War and before the Taiping Rebellion, the basic assumption still held that purchase could propel a candidate directly into a position within a manageable waiting time. By the latter half of the nineteenth-century, waiting times required longevity. Hu Heng cites a case of a certain Wang Buduan æ±ªæ­¥ç«¯ who was appointed prefect in 1906 at age 52. He had purchased the Imperial Academy degree (P11 ç£ç) and the rank of deputy prefect (P119 *tongzhi* åç¥, 5a) in 1864, at the age of 10. In 1903, he again purchased a rank elevation and a peacock-feather plume decoration (P413 *hualing* è±ç¿). In the meantime, he served as expectant official in various temporary commissions, before he was finally recommended and appointed to a vacancy after 42 years (*Gansu Xinjiang xunfu liankui zou* çèæ°çå·¡æ«è¯é­å¥ï¼GX 32.6.19, cited from <cite data-cite="14298532/NMH52TDZ"></cite>, 381-382). The existence of a growing corps of expectant officials was recognized in the later game boards in the compartment for âExpectant officials by Purchaseâ (C14 *juanban houbu* æç­åè£). The new charts added these expectancies as an additional waiting loop so that extra rolls of Virtue, Talent, and Merit were required for a posting to a vacancy. In comparison to reality, this was a very mild obstacle. Moreover, no new rules were created to buttress the purchase.

As stated above, purchase is the only part of the game, where personal strategy comes into play. Modern recreations of the game try to eliminate purchase, because it destroys suspense and is seen as a sign of corruption (<cite data-cite="14298532/4E7IWGZF"></cite>, 87-88). But in the logic of the game design, purchase does not contradict or negate meritocracy. The Imperial Academy degree, which was mostly purchased and was the precondition for all further rank purchase, is treated like a normal examination degree with the assumption that the player will go on to succeed in the imperial examinations. This also corresponded to the real behavior of many literati. Once in the âExpectantâ category, further promotion by dice is âmeritocratic.â The location of the new block is also revealing. It is placed above those imperial rewards and closer to the imperial center of the game. In a game where a career is the ultimate goal, faster promotion is closer to Heaven.
If we want to find out how a player would use strategy to win the game and whether they would be successful, we are faced with two problems: First, a player could purchase expectant rank from any position, so we have to devise a few plausible purchase scenarios. ## How much money can be earned?

Before investigating the effects of office purchase in our simulation, we still need to first examine how the distribution of money behaves without this mechanism. Recall that from the rules of the Mandarin Promotions, it is clear that it is a zero-sum game, so that the expected return for each player is precisely 0. However, this does not mean that when players would gather to play a few rounds, their subjective experience is one where they neither loose nor gain any money. In fact, we have already seen above that players that are lucky enough to get any of the highly desirable hereditary ranks stand to make a lot of money on average. However, these positions can only be gained by rolling four of the same die in the beginning of the game, so the probability is only $\frac{6}{6^4}\approx 0.0046$. So how often do players really make (or lose) a lot of money? To get an overview, we can compute the histogram of the monetary results for all players using data collected by our simulator. The result of this is displayed in [Figure 14](figure-money-histogram-*).

```python
# Get list of effective money for all playthroughs
effective_money = [s.effective_money() for s in stats]

# Plot histogram
plt.hist(effective_money, bins=60, range=(-600, 600), density=True, label="Effective money")
plt.legend()
plt.tight_layout()
```

As can be seen, most of the time, the return for players is close to zero, with the proportion of games where a player wins or loses falling very sharply with the amount of money. There are however also rarely occurring configurations in which one player wins or loses more than 500 chips. To put this into perspective, recall that initially, every player has to put 100 chips into the pool, so if one player wins more than 400 chips, this means that the other three players not only lose their 100 chips, but also have had to enter more money into the pool in the course of the game (here we decided to follow Pukâs interpolation of the rules). That very high gains are more likely than very high losses, can be explained by the fact that whenever there are payouts from the pool to one player, the loss is shared among several remaining players. On the other hand, such high gains are an exception statistically. Most of the times, as we have seen, each of the four players leaves the game with very moderate gains or losses. In conclusion, it appears that Mandarin Promotions is a very moderate form of gambling, where players would not expect to find their purse much lighter or heavier after a few rounds of play. This is confirmed in historical perspective by the fact that Mandarin Promotions did not fall under the gambling ban and was enjoyed widely in very respectable families.

### When would purchase be lucrative?

In general, the question under which circumstances purchase of office is beneficial to the player would require an extremely involved game theoretic analysis, because the strategies pursued by other players might affect the choice of whether one should buy an office or not. Hence, to tackle this difficult question, as stated above, we simply evaluate the results of office purchase under the assumption that all players act according to four strategies. For the analysis, we can then once again group the collected data into those cases where the players did purchase office, and those where they did not. Furthermore, since two of the strategies target players who find themselves in particularly unfortunate circumstances, it would not be very meaningful to compare their average results to the entirety of other playthroughs. Instead, we have added a condition so that players will randomly not purchase office even though the strategy defined by us would suggest they should buy. This way, we can directly compare the results of office purchase with players who were in a similarly unfavorable situation. 
Let us now consider the results of each of the strategies:

#### Random Purchase
Random purchase, that is, in the beginning of their turn, when the player has the chance to purchase office, they will do so with a probability of 1%, selecting the office they buy randomly among the available options. By our design, the player can purchase a position in the C14 (*juanban houbu* æç­åè£) section as long as it does not decrease their rank, so even with this relatively low probability in each single round, over all 400,000 playthroughs, there were 65,705 cases where a player did purchase office. Looking at the results of purchase in this scenario, displayed in [Figures 15-17](figure-random-moves-*) we can observe that it is not beneficial at all: the players who did purchase office did not only need more turns to reach retirement (38 vs 36.4), they also had a negative average return of money. For this scenario, the detrimental effects of purchase could be explained by the fact that the strategy implemented here is completely random, so players might even spend a significant amount of money to abandon a career that was going well. Hence, let us consider next the cases where purchase is used to get a leg up in a career not going great.

```python
# Load stats for random purchase
stats_random = []
with gzip.open("script/statistics_random0.01.txt.gz", mode="rt") as infile:
	lines = infile.readlines()
    # Each line contains a summary for one playthrough
	print(f"Read input file 'script/statistics_random0.01.txt' containing data for {len(lines)} plays")
	
	for l in lines:
		stats_random.append(PlayerStatistics.from_line(l))

# Group by purchase
stats_by_purchase = [[], []]
for s in stats_random:
	stats_by_purchase[int(s.has_purchased_office)].append(s)

# Compute averages
average_pool_money_by_purchase = list(sum(s.money_from_pool - s.money_to_pool for s in ss) / (len(ss) if len(ss) > 0 else 1) for ss in stats_by_purchase)
average_player_money_by_purchase = list(sum(s.money_from_players - s.money_to_players for s in ss) / (len(ss) if len(ss) > 0 else 1) for ss in stats_by_purchase)
average_num_moves_by_purchase = list(sum(s.num_moves for s in ss) / (len(ss) if len(ss) > 0 else 1) for ss in stats_by_purchase)
average_final_rankint_by_purchase = list(sum(s.final_rankint for s in ss) / (len(ss) if len(ss) > 0 else 1) for ss in stats_by_purchase)
```

```python
x_pos = range(1, 3)
x_ticks = [n + f" ({l})" for n, l in zip(("Did not purchase according to strategy",
                                          "Did purchase according to strategy"),
                                         (len(ss) for ss in stats_by_purchase))]

plt.bar(x_pos, average_num_moves_by_purchase, bar_width, label="Average number of moves until retirement")
plt.xticks(x_pos, x_ticks, fontproperties=font_props, rotation=-45, ha="left")
plt.legend(loc='upper center', bbox_to_anchor=(0, 1.01, 1, 0.1))
plt.tight_layout()
```

#### Late-in-rank Strategy: Purchase when the player has not reached rank y after x turns 

In this scenario ([Figures 18-23](figure-late-a-moves-*)), the player checks in the beginning of every turn whether the number of turns they took is at least x, but their rank is ranked below y. Specifically, we tested two scenarios, one with x=5 and $y=9b$, i.e. players who still did not obtain a position with at least the lowest rank 9b after 5 turns would purchase, and one with $x=10$ and $y=8a$, i.e. players who do not have a rank of at least 8a after 10 turns will purchase. The first scenario applies to roughly one in four playthroughs, and the second to ca. 14%. In the first scenario, more than two thirds of the players who are late to obtain a (high) rank are being found in the examination system, but only about one in six in the second scenario.

Looking at the effects of purchase under these conditions, as described above, we do not only compare to those in the general population of playthroughs, but also have a âcontrol groupâ of players who are likewise late in obtaining a ranked position, but did not purchase office, thus going against the presumed strategy. Among those three groups, two effects can be clearly observed: first, buying an office is in general not a financially sound decision in our simulation, and second, it does, however, speed up significantly the way to retirement in a higher rank. As can be seen from the average number of moves until retirement for the three groups, those who are late in the early game in terms of acquiring rank are also on average late to finish the game, unless they decide to purchase an office, in which case their time to retirement gets significantly shortened, in one of the two scenarios even to a value below the average over all playthroughs where the condition did not apply. On the other hand, in terms of monetary effects, in both scenarios, the high cost of buying a position means that players who do so in the scenarios lose a considerable amount of money on average. In the case of $x=5$ and $y=9b$, this is in contrast to the group who did not purchase office even though they should have according to the strategy, who have a positive average return. On the other hand, for $x=10$ and $y=8a$, those who do purchase office still have a negative average return, but the results for those who do not buy a position despite the condition applying to them, thus going against strategy are even worse.

### Conclusion

In conclusion, it appears that â at least when using our rules for office purchase â the concerns of modern authors regarding office purchase are unfounded. Our experiment has shown that arbitrary purchase of office actually has a detrimental effect on the players, and even when used under circumstances that are less than ideal, purchase can mostly just compensate for a lack of merit in the form of dice roles, but has a dubious economic value. Hence, it might actually be a more interesting choice to allow the purchase of office, because players canât be certain whether they will benefit from it or not.

## Conclusions

The popular perception of the bureaucracy in China has long been obsessed with its âhidden rulesâ (qian guize æ½è¦å) (<cite data-cite="14298532/BJI3GF7E"></cite>). Finding the hidden rules of the Qing boardgame Mandarin Promotions, we thought, may reveal some truths about how contemporary people perceived it. As it was expressed in a poem in 1887:

> It is not the [ancient game of] Chaupar, but contains a nick for satire.  
> Where the god of wealth determines demotions and transfers,  
> Where [dice made from] dried bones rule over rise and fall,  
> Where the exalted [officers] get payouts from the pool,  
> While the subordinates lose money.  
> Had I known earlier those things, I would not have entered official life.  
(<cite data-cite="14298532/KQMVEIDP"></cite>)

In fact, our digital exploration shows that the game allows a glimpse into both the historical construction of the Qing mandarinate as well as the game designerâs perception of it.

We compared two methods: the network approach and a Monte Carlo simulation. While the network approach provided insights into the personnel and appointment structure depicted in the game, only the Monte Carlo simulation with four players could conclusively answer our research questions about the distribution of opportunities and the advisability of purchase as a career strategy. The simulation revealed the hidden privileges in the Qingâs monarchic system which defy common perceptions of a rational meritocratic Qing bureaucracy. The main difficulty in deploying both methods was in implementing several special rules that break the usual flow of the game, adding a layer of complexity compared to simpler games such as the Game of the Goose.

In conclusion, we made four major observations regarding our research questions: 


Firstly, different from the common perception that Qing officials were chosen by meritocratic examinations, the game has a multitude of starting positions which represent chushen provenances or qualification for official careers which can be broadly divided into hereditary privilege, examination, incipient professionalism, and merit outside examinations (especially military). Every provenance, except for the Student of Medicine, has a fair chance to proceed to the highest rank, which maintains the suspense of the game. However, our simulation of 100,000 playthroughs with consideration of monetary returns revealed that hereditary privilege had the highest earning capacity, while examination candidates achieved meagerly positive returns. Other provenances had negative returns on average. The most important mechanism for this result is the rule that limit access to the Grand Secretary position to all Manchus and holders of only the highest examination degree. As the poem says, the highest officials receive most of the payouts.

Secondly, our study of the centrality of positions in the personnel structure showed that some positions in the bureaucracy are more pivotal than others for advancement. We limited our exploration to the prefects and the secretaries, but other positions could be scrutinized in the same manner. We cite new research on the prefect which shows that this position was an important outlet of examination graduates who waited in the underpaid metropolitan bureaucracy for an opportunity to be appointed to a lucrative provincial position. 
Thirdly, our network model in comparison with historical sources answered our question about the historicity of the personnel structure depicted in the game. We could show that its rules were based in the Rank Review (*pinjikao*) charts that governed official appointment paths. Even though our board hails from the late nineteenth century, it did not reflect changes after 1843. We can thus confirm that the unknown creator who wrote the preface of 1840 (Oxford chart) probably was the original designer of this game board and that later versions made only minor changes but did not touch the general structure of the game.

Fourthly, while Mandarin Promotions is a pure dice game, office purchase could potentially bring in an element of strategy. This feature, which was not seen in surviving late Ming and early Qing boards, has been allowed from the conception of the 1840 game board, although the rules remained incomplete. It reflects the expansion of office selling in the early decades of the nineteenth century. Modern recreations of the game in Hong Kong like that by Puk Wing Kin tend to discourage or ban office selling, because it destroys the suspense of the game. However, the simulation of 400,000 playthroughs of the game reveals that in most circumstances purchase did not reward the player if monetary gain is considered. Purchase appears mostly suited to those players who are lagging behind. as a chance to make up for their lack of good dice rolls. Since the player cannot anticipate the outcome, purchase does not diminish the suspense of the game, nor does it devalue hereditary privileges or examination degrees.
